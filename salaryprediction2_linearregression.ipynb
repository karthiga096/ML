```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
saldf=pd.read_csv('/content/drive/MyDrive/Salary Data.csv')
```

```python
saldf.head()
```

```python
saldf.info()
```

```python
saldf.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le_job = LabelEncoder()
le_education = LabelEncoder()
saldf["Job Title"] = le_job.fit_transform(saldf["Job Title"])
saldf["Education Level"] = le_education.fit_transform(saldf["Education Level"])
saldf.head()
```

```python
saldf["Job Title"]=le_job.inverse_transform(saldf["Job Title"])
saldf.head()
```

```python
inp=saldf[['Years of Experience']]
out=saldf['Salary']
```

```python
LR=LinearRegression()
```

```python
Train_data=pd.concat([inp,out],axis=1)
train=Train_data.dropna()
```

```python
inp1=train[['Years of Experience']]
out1=train['Salary']
```

```python
LR.fit(inp1,out1)
```

```python

plt.scatter(inp1,out1)
plt.plot(inp1,LR.predict(inp1))
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("salary prediction")
plt.show()
```
