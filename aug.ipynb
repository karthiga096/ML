```python
from PIL import Image, ImageEnhance, ImageFilter
import matplotlib.pyplot as plt

# Load image
img_path = "/content/drive/MyDrive/karthika/WhatsApp Image 2025-12-16 at 3.45.12 PM (1).jpeg"
img = Image.open(img_path)

# Create augmentations
augmented_images = {
    "Original": img,
    "Horizontal Flip": img.transpose(Image.FLIP_LEFT_RIGHT),
    "Rotate 15": img.rotate(15),
    "Rotate -15": img.rotate(-15),
    "Brightness Increase": ImageEnhance.Brightness(img).enhance(1.4),
    "Brightness Decrease": ImageEnhance.Brightness(img).enhance(0.6),
    "Blur": img.filter(ImageFilter.GaussianBlur(2)),
    "Sharpness": ImageEnhance.Sharpness(img).enhance(2.0),
}

# Display images and save each one separately
plt.figure(figsize=(12, 10))
for i, (title, image) in enumerate(augmented_images.items()):
    plt.subplot(3, 3, i + 1)
    plt.imshow(image)
    plt.title(title)
    plt.axis("off")
    image.save(f"{title.replace(' ', '_').replace('', 'deg')}.png") # Save each image with a clean filename

plt.tight_layout()
plt.show()
```
