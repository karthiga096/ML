```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
from PIL import Image
import numpy as np

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    shear_range=0.2
)

input_folder = "/content/drive/MyDrive/imagess/image/birds"
output_folder = "/content/drive/MyDrive/imagess/augmented_birds"

os.makedirs(output_folder, exist_ok=True)

for img_name in os.listdir(input_folder):
    img_path = os.path.join(input_folder, img_name)

    if not img_name.lower().endswith((".jpg", ".jpeg", ".png")):
        continue

    img = Image.open(img_path).convert("RGB")
    img = img.resize((224, 224))

    # Convert to float32 for saving output
    x = np.expand_dims(np.array(img).astype("float32"), axis=0)

    i = 0
    for batch in train_datagen.flow(
        x,
        batch_size=1,
        save_to_dir=output_folder,
        save_prefix="aug",
        save_format="jpg"
    ):
        i += 1
        if i >= 10:
            break

print("\n✔ Augmentation finished! Check this folder:")
print(output_folder)
```
